```python
import pandas as pd
```

```python
df = pd.read_csv("Mudit_Tripathi.csv")
df.head()
```

```python
inputs = df.drop('Drug',axis='columns')
```

```python
inputs2 = df.drop('Patient ID',axis = 'columns')
```

```python
target = df['Drug']
```

```python
from sklearn.preprocessing import LabelEncoder
le_Age = LabelEncoder()
le_Sex = LabelEncoder()
le_BP = LabelEncoder()
le_Cholesterol = LabelEncoder()
le_Drug = LabelEncoder()
```

```python
inputs['Age_n'] = le_Age.fit_transform(inputs['Age'])
inputs['Sex_n'] = le_Sex.fit_transform(inputs['Sex'])
inputs['BP_n'] = le_BP.fit_transform(inputs['BP'])
inputs['Cholesterol_n'] = le_Cholesterol.fit_transform(inputs['Cholesterol'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['Patient ID','Age','Sex','BP','Cholesterol'],axis='columns')
```

```python
inputs_n
```

# IMPLEMENTING MODEL

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```
